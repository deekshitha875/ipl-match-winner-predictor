```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv('deliveries.csv')
print("Dataset loaded:", df.shape)
```

```python
# Total runs per team
team_runs = df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)

plt.figure(figsize=(14,6))
team_runs.plot(kind='bar', color='steelblue')
plt.title('Total Runs Scored by Each IPL Team (2008-2024)')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('team_runs.png')
plt.show()
print("Top 3 run scorers:")
print(team_runs.head(3))
```

```python
wickets = df[df['is_wicket']==1].groupby('bowling_team')['is_wicket'].sum().sort_values(ascending=False)

plt.figure(figsize=(14,6))
wickets.plot(kind='bar', color='crimson')
plt.title('Total Wickets Taken by Each IPL Team (2008-2024)')
plt.xlabel('Team')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('team_wickets.png')
plt.show()
print("Top 3 wicket takers:")
print(wickets.head(3))
```

```python
top_batsmen = df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_batsmen.plot(kind='bar', color='gold')
plt.title('Top 10 IPL Batsmen by Total Runs')
plt.xlabel('Player')
plt.ylabel('Total Runs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_batsmen.png')
plt.show()
print("Top 3 batsmen:")
print(top_batsmen.head(3))
```

```python
top_bowlers = df[df['is_wicket']==1].groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_bowlers.plot(kind='bar', color='purple')
plt.title('Top 10 IPL Bowlers by Total Wickets')
plt.xlabel('Player')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_bowlers.png')
plt.show()
print("Top 3 bowlers:")
print(top_bowlers.head(3))
```

```python
# Build team strength scores
team_batting = df.groupby('batting_team')['total_runs'].mean()
team_bowling = df[df['is_wicket']==1].groupby('bowling_team')['is_wicket'].count()

# Normalize scores
team_batting_norm = (team_batting - team_batting.min()) / (team_batting.max() - team_batting.min())
team_bowling_norm = (team_bowling - team_bowling.min()) / (team_bowling.max() - team_bowling.min())

# Overall strength
team_strength = (team_batting_norm + team_bowling_norm) / 2
team_strength = team_strength.sort_values(ascending=False)

plt.figure(figsize=(14,6))
team_strength.plot(kind='bar', color='teal')
plt.title('IPL Team Overall Strength Score')
plt.xlabel('Team')
plt.ylabel('Strength Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('team_strength.png')
plt.show()
```

```python
def predict_winner(team1, team2):
    t1_score = team_strength.get(team1, 0)
    t2_score = team_strength.get(team2, 0)

    total = t1_score + t2_score
    if total == 0:
        print("Teams not found!")
        return

    t1_prob = (t1_score / total) * 100
    t2_prob = (t2_score / total) * 100

    print(f"\n🏏 Match Prediction")
    print(f"{'='*35}")
    print(f"{team1}: {t1_prob:.1f}% win probability")
    print(f"{team2}: {t2_prob:.1f}% win probability")
    print(f"{'='*35}")

    winner = team1 if t1_score > t2_score else team2
    print(f"🏆 Predicted Winner: {winner}")

# Test predictions
predict_winner("Mumbai Indians", "Chennai Super Kings")
predict_winner("Royal Challengers Bangalore", "Kolkata Knight Riders")
```
